## Algoritmo K-means

K means es un algoritmo de clasificación no supervisada. Este algoritmo permite clasificar un conjunto de datos en K grupos, basándose en sus características.

Este algoritmo resuelve un problema de optimización, donde lo que se busca es minimizar la suma de las distancias cuadráticas de cada objeto al centroide de su cluster

La función objetivo del problema de optimización de k-means se puede escribir de la siguiente manera:

$
\min_{\mu_i} \sum_{i=1}^k \sum_{x_j \in S_i} \|x_j - \mu_i\|^2
$

Donde:
- $ \mu_i $ representa el centroide del grupo $ i $.
- $ x_j $ representa un punto de datos perteneciente al grupo $ S_i $.
- $ S_i $ es el conjunto de puntos asignados al centroide $ \mu_i $.
- $ k $ es el número de clusters.

El objetivo es minimizar la suma de las distancias cuadradas entre los puntos de datos y sus respectivos centroides.


```python
from sklearn.cluster import KMeans

df = pd.read_csv('kmeans_data.csv')


X = df[['x1', 'x2']]

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Obtener las etiquetas de los clusters
df['cluster_pred'] = kmeans.labels_

# Visualizar los clusters predichos
plt.scatter(df['x1'], df['x2'], c=df['cluster_pred'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroides')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Clusters Predichos por K-Means')
plt.legend()
plt.show()
```
