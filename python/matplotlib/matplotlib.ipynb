```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

## Para trabajar con Matplotlib requerimos un dataset

```python
data = pd.read_csv('sales.csv')
data.head()
```

## Siempre que trabajamos con fechas primero convertimos estas a formato date time. 

```python
data['Date'] = pd.to_datetime(data['Date'])
```

## Gr√°fico de eje Y versus eje X.

```python
plt.figure(figsize=(10, 6))
for product in data['Product'].unique():
    product_data = data[data['Product'] == product]
    plt.plot(product_data['Date'], product_data['Sales'], marker='o', label=product)

plt.title('Sales Over Time') #Agrega titulo al grafico
plt.xlabel('Date') #Agrega titulo al eje x
plt.ylabel('Sales') #Agrega titulo al eje y
plt.legend() #Agrega la leyenda al grafico
plt.grid(True) #Agrega grilla al grafico
plt.show()
```

## Gr√°fico de barras

```python

total_sales_by_product = data.groupby('Product')['Sales'].sum()
total_sales_by_product.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

## Gr√°fico de torta (Pie Chart)

```python
sales_by_category = data.groupby('Category')['Sales'].sum()
sales_by_category.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.ylabel('')
plt.show()
```

## Gr√°fico de Bigotes (Box Plot)

```python
plt.figure(figsize=(10, 6))
data[['Sales', 'Profit']].boxplot()
plt.title('Box Plot of Sales and Profit')
plt.ylabel('Values')
plt.show()
```

## Funci√≥n acumulada (Empirical Cumulative Distribution Function)

```python
# Extract the Sales data
sales = data['Sales']

# Calculate the ECDF values
x = np.sort(sales)
y = np.arange(1, len(x) + 1) / len(x)

# Plot the ECDF
fig, ax = plt.subplots(figsize=(10, 6))
ax.step(x, y, where='post')
ax.set_title('ECDF of Sales')
ax.set_xlabel('Sales')
ax.set_ylabel('ECDF')
plt.grid(True)
plt.show()
```
