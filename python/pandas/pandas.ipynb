```python
import pandas as pd
```

## Cargar archivos y convertir a otro formato.

## Operaciones

```python
df_join = pd.read_csv('data_join.csv')  # Abre archivo csv
df_join.sample(5)  # Muestra 5 filas aleatorias
```

## Generar un dataset Global con la información de ambas tablas (Usamos merge y tomamos como referencia la columna First Name)

```python
df_junto = pd.merge(df_csv, df_join, on='First Name')  # Hace un join de los dos dataframes
df_junto.sample(5)  # Muestra 5 filas aleatorias
```

## Eliminar columnas

Notamos que tenemos una columna repetida (Last name)

```python
df_junto = df_junto.drop(columns=['Last Name_y'])  # Elimina columnas
df_junto.sample(5)  # Muestra 5 filas aleatorias
```

## Renombrar columna

```python
df_junto = df_junto.rename(columns={'Last Name_x': 'Last Name'})  # Renombra columnas
df_junto.sample(5)  # Muestra 5 filas aleatorias
```

## Generar una nueva columna

```python
df_junto['monthly_income'] = df_junto['Salary'] / 12  # Crea una nueva columna
df_junto.sample(5)  # Muestra 5 filas aleatorias
```

## Mostrar solo algunas columnas

```python
df_junto[['First Name', 'Last Name', 'Salary']]  # Muestra solo las columnas seleccionadas
```

## Filtrar por nombre

```python
df_junto.loc[df_junto['First Name'] == 'Jack']  # Filtra por nombre
```

## Generar estadísticas del dataset

```python
df_junto.describe()  # Muestra estadísticas
```

## Eliminar filas duplicadas.

```python
df_junto = df_junto.drop_duplicates()  # Elimina duplicados
```

## Agregar una fila al final del dataset

```python
df_junto.loc[len(df_junto)] = ['John', 'Doe', "example@example.cl", "La moneda 123", "869123","Chile", "Futbolista", "ANFP", 92000, "1996-02-02", 7666.666667]  # Agrega una fila
df_junto.tail(1)
```

## Elimina todas las filas en donde haya al menos un valor nulo.

```python
df_junto = df_junto.dropna()  # Elimina filas con valores nulos
```

## Guardar el archivo final modificado.

```python
df_junto.to_csv('data_final.csv', index=False)  # Guarda en csv
```
